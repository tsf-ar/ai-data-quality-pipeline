```python
df["clean_text"] = (
    df["text"]
    .str.replace(r"[!؟،,.]+", "", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df[["text", "clean_text"]]
```

```python
allowed_labels = {"neutral", "positive", "unsafe"}
df[~df["label"].isin(allowed_labels)]
```
