```python
import pandas as pd

df = pd.read_csv("../data/cleaned_dataset.csv")
df
```

```python
df["text_length"] = df["clean_text"].str.len()

df[df["text_length"] < 10]
```

```python
df[df.duplicated(subset="clean_text", keep=False)]
```

```python
import re

def contains_english(text):
    return bool(re.search(r"[a-zA-Z]", text))

df[
    (df["language"].isin(["ur", "ar"])) &
    (df["clean_text"].apply(contains_english))
]
```

```python
df["label"].value_counts(normalize=True)
```

```python
qa_summary = {
    "total_records": len(df),
    "short_texts": (df["text_length"] < 10).sum(),
    "duplicates": df.duplicated(subset="clean_text").sum(),
}

qa_summary
```
