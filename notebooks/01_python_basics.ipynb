## Variables and Strings

Variables store information.
In AI data work, this usually means text, language codes, and labels.


```python
sentence = "میں پائتھن سیکھ رہا ہوں"
language = "ur"
label = "neutral"

print(sentence)
print(language)
print(label)
```

### Why this matters in AI data work

Each variable represents:
- a text sample
- its metadata
- its annotation label

This structure is the atomic unit of AI training data.


```python
sentences = [
    "میں پائتھن سیکھ رہا ہوں",
    "أحب البرمجة",
    "I am learning Python"
]

for s in sentences:
    print(len(s), s)
```

```python
dataset = [
    {
        "text": "میں پائتھن سیکھ رہا ہوں",
        "language": "ur",
        "label": "neutral"
    },
    {
        "text": "أحب البرمجة",
        "language": "ar",
        "label": "positive"
    }
]

for item in dataset:
    print(item["language"], item["label"], item["text"])
```

## What I learned

- Python handles Arabic and Urdu text natively
- Lists represent datasets
