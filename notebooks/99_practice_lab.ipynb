```python
import csv
import re

dataset = []
with open("../data/practice_dataset.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        dataset.append(row)
        
print(len(dataset))
```

```python
print(dataset[0]["text"])
print(dataset[0]["language"])
print(dataset[0]["label"])
```

```python
for item in dataset:
    text = item["text"]
    text = text.strip()
    text = " ".join(text.split())
    text = re.sub(r"[!؟،,.]+", "", text)
    item["clean_text"] = text
    print(text)
```

```python
for item in dataset:
    print(item["clean_text"])
```

```python
for item in dataset:
    print(len(item["clean_text"]), "|", item["clean_text"])
```

```python
for item in dataset:
    if len(item["clean_text"]) < 10:
        print("Too short:", item["clean_text"])
```

```python
for item in dataset:
    if "clean_text" not in item or "label" not in item:
        print("invalid record:", item)
```
